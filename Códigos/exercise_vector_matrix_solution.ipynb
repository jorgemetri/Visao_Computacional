```python
import matplotlib.pyplot as plt
import numpy as np

# Define the area where the plots will be displayed
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Define the geometric points and transformations
p = np.array([2, 0, 1])  # Point in homogeneous coordinates [x, y, 1]
T = np.array([[1, 0, 0], [0, 1, 2], [0, 0, 1]])  # Translation matrix
R = np.array([[np.cos(np.pi/4), -np.sin(np.pi/4), 0], 
              [np.sin(np.pi/4), np.cos(np.pi/4), 0], 
              [0, 0, 1]])  # Rotation matrix in homogeneous coordinates

# Apply transformations
p1 = T @ p  # Translated point
p2 = R @ p  # Rotated point
p3 = R @ (T @ p)  # Translated then rotated point

# Define plot settings for each subplot
for i in range(2):
    for j in range(2):
        ax[i, j].grid(True)
        ax[i, j].axhline(0, color='black', linewidth=0.75)
        ax[i, j].axvline(0, color='black', linewidth=0.75)
        ax[i, j].set_xlabel("X-axis")
        ax[i, j].set_ylabel("Y-axis")
        ax[i, j].set_xlim(-5, 5)
        ax[i, j].set_ylim(-5, 5)

# Plot original point p in ax[0, 0]
ax[0, 0].scatter(p[0], p[1], color='red', s=100, label=f'p ({p[0]:.1f},{p[1]:.1f})', zorder=5)
ax[0, 0].text(p[0] + 0.2, p[1] + 0.2, f't=0', color='red')
ax[0, 0].set_title("Original Point")
ax[0, 0].legend()

# Plot translated point p1 in ax[0, 1]
ax[0, 1].scatter(p1[0], p1[1], color='blue', s=100, label=f'p1 ({p1[0]:.1f},{p1[1]:.1f})', zorder=5)
ax[0, 1].text(p1[0] + 0.2, p1[1] + 0.2, f't=1', color='blue')
ax[0, 1].set_title("Translated Point")
ax[0, 1].legend()

# Plot rotated point p2 in ax[1, 0]
ax[1, 0].scatter(p2[0], p2[1], color='green', s=100, label=f'p2 ({p2[0]:.2f},{p2[1]:.2f})', zorder=5)
ax[1, 0].text(p2[0] + 0.2, p2[1] + 0.2, f't=3', color='green')
ax[1, 0].set_title("Rotated Point")
ax[1, 0].legend()

# Plot translated then rotated point p3 in ax[1, 1]
ax[1, 1].scatter(p3[0], p3[1], color='purple', s=100, label=f'p3 ({p3[0]:.2f},{p3[1]:.2f})', zorder=5)
ax[1, 1].text(p3[0] + 0.2, p3[1] + 0.2, f'p3 ({p3[0]:.2f},{p3[1]:.2f})', color='purple')
ax[1, 1].set_title("Translated then Rotated Point")
ax[1, 1].legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

Exercise: What do you need to do to move the object so point A stays at point P1, and then rotates by 45 ° with respect to point P2? Define the transformation arrays and the order in which they should be applied.

```python
# Define the object using homogeneous coordinates

obj = np.array([[6.5, 7, 1],[8.5, 5, 1],[6.5, 5, 1],[6, 3.5, 1],[5, 4.5, 1],[6.5, 7, 1]])
obj = obj.T

obj_ori = obj

print('obj=\n', obj)

# Plot the object
fig = plt.figure()
ax0 = plt.axes()
ax0.grid()
ax0.set_title("Rigid Body Tranformations")
ax0.plot(obj[0,:],obj[1,:],'gray')
ax0.text(obj[0,0]+0.1,obj[1,0]+0.1,"original object")
ax0.set_xlim([-10,10])
ax0.set_ylim([-10,10])
ax0.set_aspect('equal')

# Define P1 and P2

P1 = np.array([2, 3.5, 1])
P2 = np.array([5, 1, 1])

# Plot P1 and P2
ax0.plot(P1[0],P1[1],'*r')
ax0.text(P1[0]+0.1,P1[1]+0.1,"P1")
ax0.plot(P2[0],P2[1],'*b')
ax0.text(P2[0]+0.1,P2[1]+0.1,"P2")
ax0.plot(0,0,'*')

# For each transformation needed, define:
# 1- the transformation matrix
# 2- apply the transformation to the object
# 3- plot the current result

T1 = np.array([[1, 0, -4.5],[0, 1, -3.5],[0, 0, 1]])
obj = T1@obj
ax0.plot(obj[0,:],obj[1,:],'red')
ax0.text(obj[0,0]+0.1,obj[1,0]-0.5,"step 1")

T2 = np.array([[1, 0, -P2[0]],[0, 1, -P2[1]],[0, 0, 1]])
obj = T2@obj
ax0.plot(obj[0,:],obj[1,:],'blue')
ax0.text(obj[0,0]+0.1,obj[1,0]+0.1,"step 2")

R = np.array([[np.cos(pi/4), -np.sin(pi/4),0],[np.sin(pi/4), np.cos(pi/4),0],[0, 0, 1]])
obj = R@obj
ax0.plot(obj[0,:],obj[1,:],'green')
ax0.text(obj[0,0]+0.1,obj[1,0]+0.1,"step 3")

T3 = np.array([[1, 0, P2[0]],[0, 1, P2[1]],[0, 0, 1]])
obj = T3@obj
ax0.plot(obj[0,:],obj[1,:],'orange')
ax0.text(obj[0,0]+0.1,obj[1,0]+0.1,"step 4")

# Define the complete transformation by multiplying all the transformation matrices

M = T3 @ R @ T2 @ T1

# Apply the composed transformation to the original object and compare the final
# result to the previous result obtained from the step by step process

obj2 = M@obj_ori

print(obj)
print(obj2)

# Plot the original object and final result

fig = plt.figure()
ax1 = plt.axes()
ax1.grid()
ax1.set_title("Original and Final object")
ax1.plot(obj_ori[0,:],obj_ori[1,:],'gray')
ax1.set_xlim([-10,10])
ax1.set_ylim([-10,10])
ax1.set_aspect('equal')

ax1.plot(obj2[0,:],obj2[1,:],'orange')
```
